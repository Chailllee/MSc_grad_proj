```python
import pandas as pd
import os
```

First, let's split the StationFootfall_2023_2024.csv into two separate files: StationFootfall_2023.csv and StationFootfall_2024.csv.

```python
# Path to the input file
input_file = 'D:\File_auto\\0_UCL_CASA\OneDrive - University College London\Xiaoyi_dissertation\Analysis\Data\\tfl_crowding_data\StationFootfall_2023_2024 .csv'

# Read the input file
df = pd.read_csv(input_file)

# Convert 'TravelDate' to datetime format
df['TravelDate'] = pd.to_datetime(df['TravelDate'], format='%Y%m%d')

# Split into two DataFrames for 2023 and 2024
df_2023 = df[df['TravelDate'].dt.year == 2023]
df_2024 = df[df['TravelDate'].dt.year == 2024]

# Save the DataFrames to separate CSV files
df_2023.to_csv('D:\\File_auto\\0_UCL_CASA\\OneDrive - University College London\\Xiaoyi_dissertation\\Analysis\\Data\\tfl_crowding_data\\StationFootfall_2023.csv', index=False)
df_2024.to_csv('D:\\File_auto\\0_UCL_CASA\\OneDrive - University College London\\Xiaoyi_dissertation\\Analysis\\Data\\tfl_crowding_data\\StationFootfall_2024.csv', index=False)

print("Files split and saved successfully.")
```

# Step 1: Extract Specific Stations and Add Borough Information

```python
# Define the stations for each borough
borough_stations = {
    'City of London': ['Aldgate', 'Bank', 'Barbican', 'Blackfriars', 'Cannon Street', 'Liverpool Street', 
                       'Mansion House', 'Monument', 'Moorgate', 'St Paul\'s', 'Chancery Lane'],
    'Westminster': ['Baker Street', 'Bayswater', 'Bond Street', 'Charing Cross', 'Covent Garden', 'Edgware Road', 
                    'Embankment', 'Great Portland Street', 'Green Park', 'Hyde Park Corner', 'Lancaster Gate', 
                    'Leicester Square', 'Maida Vale', 'Marble Arch', 'Marylebone', 'Oxford Circus', 'Paddington', 
                    'Piccadilly Circus', 'Pimlico', 'Queensway', 'Regent\'s Park', 'Royal Oak', 'St James\'s Park', 
                    'St John\'s Wood', 'Temple', 'Victoria', 'Warwick Avenue', 'Westbourne Park', 'Westminster'],
    'Camden': ['Belsize Park', 'Camden Town', 'Chalk Farm', 'Euston', 'Euston Square', 'Finchley Road', 'Goodge Street',
               'Hampstead', 'Holborn', 'Kentish Town', 'King\'s Cross St Pancras', 'Mornington Crescent', 'Russell Square', 
               'Swiss Cottage', 'Tottenham Court Road', 'Warren Street', 'West Hampstead'],
    'Islington': ['Angel', 'Archway', 'Arsenal', 'Caledonian Road', 'Farringdon', 'Finsbury Park', 'Highbury & Islington', 
                  'Holloway Road', 'Tufnell Park', 'Old Street'],
    'Kensington and Chelsea': ['Earl\'s Court', 'Gloucester Road', 'High Street Kensington', 'Holland Park', 'Kensington (Olympia)', 
                               'Knightsbridge', 'Ladbroke Grove', 'Latimer Road', 'Notting Hill Gate', 'Sloane Square', 
                               'South Kensington', 'West Brompton'],
    'Lambeth': ['Brixton', 'Clapham Common', 'Clapham North', 'Lambeth North', 'Oval', 'Stockwell', 'Vauxhall', 'Waterloo'],
    'Southwark': ['Bermondsey', 'Borough', 'Canada Water', 'Elephant & Castle', 'Kennington', 'London Bridge', 'Southwark']
}

# Directory containing the footfall data
data_dir = 'D:\\File_auto\\0_UCL_CASA\\OneDrive - University College London\\Xiaoyi_dissertation\\Analysis\\Data\\tfl_crowding_data'

# List of footfall data files
file_paths = [
    'StationFootfall_2019.csv',
    'StationFootfall_2020.csv',
    'StationFootfall_2021.csv',
    'StationFootfall_2022.csv',
    'StationFootfall_2023.csv',
    'StationFootfall_2024.csv'
]

# Directory to save the filtered files
filtered_dir = data_dir

# Loop through each file and filter stations
for file_path in file_paths:
    full_path = os.path.join(data_dir, file_path)
    df = pd.read_csv(full_path)
    
    # Initialize an empty DataFrame to hold the filtered data
    filtered_data = pd.DataFrame()
    
    for borough, stations in borough_stations.items():
        borough_df = df[df['Station'].isin(stations)]
        borough_df['Borough'] = borough
        filtered_data = pd.concat([filtered_data, borough_df])
    
    # Save the filtered data
    filtered_file_path = os.path.join(filtered_dir, 'filtered_' + os.path.basename(file_path))
    filtered_data.to_csv(filtered_file_path, index=False)
    print(f"Filtered data saved to {filtered_file_path}")
```

# Step 2: Merge Filtered Data with Station Information

```python
# Path to the station information file
station_info_path = 'D:\\File_auto\\0_UCL_CASA\\OneDrive - University College London\\Xiaoyi_dissertation\\Analysis\\Data\\tfl_geo_data\\Stations_20180921.csv'

# Read the station information
station_info_df = pd.read_csv(station_info_path)

# Directory containing the filtered footfall data
filtered_dir = 'D:\\File_auto\\0_UCL_CASA\\OneDrive - University College London\\Xiaoyi_dissertation\\Analysis\\Data\\tfl_crowding_data'

# Loop through each filtered file and merge with station information
for file_path in os.listdir(filtered_dir):
    if file_path.startswith('filtered_') and file_path.endswith('.csv'):
        filtered_df = pd.read_csv(os.path.join(filtered_dir, file_path))
        
        # Merge with station information
        merged_df = pd.merge(filtered_df, station_info_df[['NAME', 'LINES', 'NETWORK', 'Zone', 'x', 'y']], left_on='Station', right_on='NAME', how='left')
        
        # Drop the 'NAME' column from the merged DataFrame
        merged_df.drop(columns=['NAME'], inplace=True)
        
        # Save the merged data
        merged_file_path = os.path.join(filtered_dir, file_path.replace('.csv', '_merged.csv'))
        merged_df.to_csv(merged_file_path, index=False)
        print(f"Merged data saved to {merged_file_path}")
```
